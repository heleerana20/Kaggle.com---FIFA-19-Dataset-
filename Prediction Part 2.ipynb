# Feature Selection

```python
features = ['International Reputation','Crossing','Finishing','ShortPassing','FKAccuracy', 'Balance','Strength','StandingTackle', 'SlidingTackle', 'Overall', 'Weak Foot','Skill Moves','Potential', 'Composure', ]
fifa_dataset = fifa_raw_dataset[[*features]]
fifa_dataset.head()
```

```python
fifa_dataset.shape
```

# Data preprocessing

```python
fifa_dataset['International Reputation'].fillna(fifa_dataset['International Reputation'].mean(), inplace = True)
fifa_dataset['Crossing'].fillna(fifa_dataset['Crossing'].mean(), inplace = True)
fifa_dataset['Finishing'].fillna(fifa_dataset['Finishing'].mean(), inplace = True)
fifa_dataset['ShortPassing'].fillna(fifa_dataset['ShortPassing'].mean(), inplace = True)
fifa_dataset['FKAccuracy'].fillna(fifa_dataset['FKAccuracy'].mean(), inplace = True)
fifa_dataset['Balance'].fillna(fifa_dataset['Balance'].mean(), inplace = True)
fifa_dataset['Strength'].fillna(fifa_dataset['Strength'].mean(), inplace = True)
fifa_dataset['Crossing'].fillna(fifa_dataset['Crossing'].mean(), inplace = True)
fifa_dataset['StandingTackle'].fillna(fifa_dataset['StandingTackle'].mean(), inplace = True)
fifa_dataset['SlidingTackle'].fillna(fifa_dataset['SlidingTackle'].mean(), inplace = True)
fifa_dataset['Composure'].fillna(fifa_dataset['Composure'].mean(), inplace = True)
fifa_dataset['Overall'].fillna(fifa_dataset['Overall'].mean(), inplace = True)
fifa_dataset['Weak Foot'].fillna(fifa_dataset['Weak Foot'].mean(), inplace = True)
fifa_dataset['Skill Moves'].fillna(fifa_dataset['Skill Moves'].mean(), inplace = True)
fifa_dataset['Potential'].fillna(fifa_dataset['Potential'].mean(), inplace = True)
```

```python
df = fifa_dataset.astype(int)
# df = df["col1"].astype(int)
```
